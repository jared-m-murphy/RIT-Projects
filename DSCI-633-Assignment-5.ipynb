### Problem 2 (6 points)

For the [figure shown](https://drive.google.com/file/d/1CNLNMvE7paIddweb2si2mtpUTHy-pEg6/view?usp=sharing), describe in a few words, the clusters you may get if you were to use a center-based, contiguity based and a density-based clustering technique.

### Solution

A center based clustering places each point in a cluster corresponding to its closest centroid, so there is most naturally going to be three globular clusters, one surrouding the two lines in the bottom left, one surrounding the small glob in the bottom right, and one surrounding the entire bullseye figure at the top.  With contiguity based clustering, where a cluster is a set of points that are most similar/closest to one , the clustering will be less globular but will extend out from their defined shapes to accomodate for the noise.  For density based clustering the noisy background will be filtered out and all clusters should naturally be defined as the two lines in the bottom-left, the small glob on the bottom right, the bullseye and the circle around it.
